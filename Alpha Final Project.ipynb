### **Metric Evaluation**

| **ACTUAL / PREDICTED** | **Not Churn** | **Churn** |
|--------------------------|---------------|------------|
| **Not Churn** | **True Negative (TN)**<br>Model predicts customer will not churn, and indeed does not churn. | **False Positive (FP)**<br>Model predicts customer will churn, but actually does not churn. |
| **Churn** | **False Negative (FN)**<br>Model predicts customer will not churn, but actually churns. | **True Positive (TP)**<br>Model predicts customer will churn, and indeed truly churns. |

Penjelasan Kesalahan dan Konsekuensinya
- Type 1 Error (False Positive)

Konsekuensi bisnis: sumber daya marketing seperti promosi, diskon, atau cashback terpakai sia-sia untuk pelanggan yang sebenarnya tidak akan pergi.
- Type 2 Error (False Negative)

Konsekuensi bisnis: perusahaan kehilangan pelanggan potensial karena tidak dilakukan tindakan pencegahan (misalnya pengiriman promo atau reminder).

---


Berdasarkan konsekuensi kesalahan prediksi, terutama karena kehilangan pelanggan (*False Negative*) memiliki dampak bisnis yang jauh lebih besar dibandingkan memberikan promo kepada pelanggan yang sebenarnya tidak akan churn (*False Positive*), maka model harus mampu mendeteksi pelanggan berisiko churn dengan sangat baik, artinya **Recall** harus menjadi prioritas utama.

Untuk mencapai tujuan tersebut, seluruh proses evaluasi model, mulai dari pemilihan model (*model selection*) hingga penentuan threshold, menggunakan **F2-Score** sebagai metrik utama. F2-Score memberikan bobot dua kali lebih besar pada Recall dibandingkan Precision, sehingga relevan dengan kebutuhan bisnis.

**Rumus Fβ-Score dan F2-Score**

Fβ-Score didefinisikan sebagai:

$$
F_{\beta} = (1 + \beta^2) \times \frac{(\text{Precision} \times \text{Recall})}{(\beta^2 \times \text{Precision}) + \text{Recall}}
$$

Ketika β = 2, Recall diberi bobot empat kali lebih besar dibanding Precision:

$$
F_{2} = 5 \times \frac{(\text{Precision} \times \text{Recall})}{(4 \times \text{Precision}) + \text{Recall}}
$$

Dengan demikian, F₂-Score memastikan bahwa model lebih fokus dalam menangkap pelanggan yang benar-benar akan churn, meskipun harus mengorbankan sedikit Precision.

**1. F2-Score untuk Model Selection**

Dalam tahap pemilihan model, F2-Score digunakan untuk mengevaluasi kemampuan setiap model dalam mengidentifikasi pelanggan churn secara sensitif.

Alasan penggunaan F2-Score:

* Menekankan Recall untuk meminimalkan *False Negative* (kehilangan pelanggan).
* Tetap mempertimbangkan Precision, agar jumlah *False Positive* tidak berlebihan.
* Metrik langsung relevan dengan tujuan bisnis, tanpa perlu metrik lain yang kurang fokus pada risiko churn.

**Contoh:**
Jika Logistic Regression memiliki F2-Score = 0.64 dan XGBoost = 0.72, maka XGBoost lebih unggul karena lebih mampu mendeteksi pelanggan berisiko churn.

**2. F2-Score untuk Threshold Optimization**

Setelah model terbaik terpilih, penentuan threshold juga menggunakan F2-Score untuk memilih ambang probabilitas yang menghasilkan keseimbangan terbaik antara Recall dan Precision, dengan penekanan kuat pada Recall.

* Threshold optimal adalah threshold yang menghasilkan F2-Score tertinggi.
* Threshold yang lebih rendah biasanya meningkatkan Recall, sehingga mengurangi risiko pelanggan churn terlewat (*False Negative*).
* Peningkatan *False Positive* masih dapat ditoleransi karena dampak bisnisnya lebih kecil.

**Contoh:**

* Threshold 0.5 → F2-Score = 0.68
* Threshold 0.4 → F2-Score = 0.76

Dengan demikian, threshold 0.4 lebih ideal karena meningkatkan sensitivitas model dalam mendeteksi churn.

Dengan berfokus pada F2-Score, model yang dikembangkan tidak hanya memiliki performa prediktif yang kuat, tetapi juga lebih selaras dengan kebutuhan bisnis. Pendekatan ini membantu perusahaan mengidentifikasi pelanggan berisiko churn secara lebih akurat, memaksimalkan upaya retensi, serta mengoptimalkan penggunaan biaya marketing agar lebih efisien.

```python
# Check detail outliers

def find_anomalies(df_model, column):
    q1 = df_model[column].quantile(0.25)
    q3 = df_model[column].quantile(0.75)
    iqr = q3 - q1
    limit = iqr * 1.5

    limit_bawah = q1 - limit
    limit_atas = q3 + limit

    print(f'[{column}]')
    print(f'Q1: {q1:.2f}')
    print(f'Q3: {q3:.2f}')
    print(f'IQR: {iqr:.2f}')
    print(f'Lower Bound: {limit_bawah:.2f}')
    print(f'Upper Bound: {limit_atas:.2f}')

    # Hitung jumlah outliers
    num_outliers = df_model[(df_model[column] < limit_bawah) | (df_model[column] > limit_atas)].shape[0]
    print(f'Jumlah outlier: {num_outliers}')
    print('-' * 40)

cols = ['Tenure', 'CityTier','WarehouseToHome','HourSpendOnApp',
        'NumberOfDeviceRegistered','SatisfactionScore','NumberOfAddress', 'Complain',
        'OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder', 
        'CashbackAmount'
        ]

for col in cols:
    find_anomalies(df_model, col)
```

Setelah melakukan eksplorasi pada variabel numerik, langkah berikutnya adalah menganalisis bagaimana variabel-variabel kategorikal berhubungan dengan perilaku churn.

```python
from scipy.stats import chi2_contingency

cat_cols = df_model[['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus']]
chi2_results = []

for col in cat_cols:
    if col != 'Churn':
        table = pd.crosstab(df_model[col], df_model['Churn'])
        
        chi2, p, dof, expected = chi2_contingency(table)
        
        chi2_results.append({
            'Feature': col,
            'Chi2 Statistic': chi2,
            'p-value': p,
            'Decision': 'Reject H₀ (Significant)' if p < 0.05 else 'Fail to Reject H₀ (Not Significant)'
        })

chi2_df = pd.DataFrame(chi2_results).sort_values(by='p-value')
chi2_df
```

**Korelasi Fitur Kategorikal dengan Churn**

Berdasarkan hasil uji Chi-Square menunjukkan bahwa setiap fitur kategorikal dari Dataset E-commerce tersebut memiliki hubungan yang signifikan secara statistik.
- H₀ (Hipotesis nol) → Tidak ada hubungan antara variabel kategorikal dan churn.
- H₁ (Hipotesis alternatif) → Ada hubungan antara variabel kategorikal dan churn.
Jika nilai p-value < 0.05, maka kita menolak H₀, artinya fitur tersebut berhubungan signifikan dengan perilaku churn pelanggan.

Setiap model memiliki karakteristik dan cara kerja yang berbeda, sehingga membantu memberikan gambaran yang lebih menyeluruh terhadap pola pada data. Pendekatan ini memastikan bahwa pemilihan model akhir dilakukan secara objektif berdasarkan evaluasi yang konsisten.

**1. Logistic Regression (LogReg)** : model linear yang memprediksi probabilitas kelas menggunakan fungsi logit. Cocok sebagai baseline karena sederhana, cepat, mudah diinterpretasikan, dan memberi gambaran awal tentang hubungan linear antar fitur.

**2. K-Nearest Neighbors (KNN)** : model non-parametrik yang melakukan klasifikasi berdasarkan kedekatan jarak dengan beberapa tetangga terdekat. Baik untuk pola non-linear dan tidak membutuhkan proses training, tetapi sensitif terhadap skala fitur dan kurang efisien pada data besar.

**3. Decision Tree Classifier** : membangun struktur pohon keputusan dengan memilih fitur yang paling informatif pada setiap percabangan. Mudah dipahami dan menangkap pola non-linear, namun rentan overfitting tanpa pengaturan parameter.

**4. Random Forest Classifier** : metode ensemble yang menggabungkan banyak decision tree secara acak. Lebih stabil, tahan terhadap overfitting, dan mampu memberikan performa yang konsisten pada berbagai jenis data.

**5. AdaBoost Classifier** : algoritma boosting yang memberi bobot lebih besar pada sampel yang salah diprediksi untuk meningkatkan kinerja weak learner. Efektif untuk pola yang sulit dipisahkan dan sering meningkatkan akurasi.

**6. Gradient Boosting Classifier** : teknik boosting yang membangun model secara bertahap dengan meminimalkan error residual. Mampu menangkap interaksi fitur dan pola kompleks, walaupun membutuhkan tuning yang lebih hati-hati.

**7. XGBoost Classifier** : implementasi gradient boosting yang dioptimalkan untuk kecepatan dan efisiensi. Memiliki regularisasi kuat, dapat menangani missing value, dan unggul pada data kompleks maupun tidak seimbang.

**8. LightGBM Classifier** : algoritma boosting berbasis tree dengan strategi leaf-wise growth. Sangat cepat, efisien untuk dataset besar, dan memiliki kemampuan generalisasi yang baik, menjadikannya salah satu model boosting paling kompetitif.

```python
# Define resampler
smote = SMOTE(random_state=0)
ros = RandomOverSampler(random_state=0)
nmiss = NearMiss(version=1)
rus = RandomUnderSampler(random_state=0) 

resamplers = [smote, ros, nmiss, rus]
```

Selanjutnya, karena data yang digunakan merupakan **data imbalance**, di mana jumlah sampel pada kelas mayoritas jauh lebih besar dibanding kelas minoritas, kondisi ini dapat menyebabkan model bias terhadap kelas mayoritas dan menurunkan kemampuan model dalam mengenali kelas minoritas (churn). Untuk mengatasi hal tersebut, digunakan beberapa teknik *resampling* sebagai berikut:

1. **SMOTE (Synthetic Minority Over-sampling Technique)** : teknik *oversampling* yang menghasilkan contoh baru pada kelas minoritas melalui proses interpolasi antar sampel yang berdekatan. Metode ini meningkatkan proporsi kelas minoritas tanpa sekadar menduplikasi data. Teknik ini digunakan karena mampu menangkap variasi data minoritas secara lebih baik dan sering memberikan peningkatan performa model pada data yang sangat tidak seimbang.

2. **Random Oversampler** : teknik *oversampling* sederhana yang bekerja dengan menggandakan sampel kelas minoritas secara acak hingga proporsinya lebih seimbang. Metode ini digunakan karena memberikan baseline yang mudah dipahami, cepat diimplementasikan, serta membantu meningkatkan sensitivitas model terutama pada kelas minoritas.

3. **Random UnderSampler** : teknik *undersampling* yang menyeimbangkan data dengan mengurangi jumlah sampel kelas mayoritas secara acak. Teknik ini digunakan karena dapat mempercepat proses pelatihan, mengurangi dominasi kelas mayoritas, serta membantu mengevaluasi performa model ketika data dibuat lebih ringkas.

4. **NearMiss** : teknik *undersampling* yang memilih sampel kelas mayoritas berdasarkan jarak terdekat terhadap sampel kelas minoritas. Dengan mempertahankan sampel yang paling relevan secara spasial, teknik ini membantu model membentuk batas keputusan yang lebih jelas. NearMiss digunakan karena mampu menjaga struktur hubungan antar kelas, terutama saat perbedaan distribusi terlalu besar.

**Model Benchmarking**

Pada tahap ini, seluruh kombinasi model dan teknik resampling dievaluasi menggunakan pipeline yang berisi preprocessing, resampling (SMOTE, RandomOverSampler, RandomUnderSampler dan NearMiss), dan algoritma machine learning. Proses evaluasi dilakukan dengan StratifiedKFold cross-validation untuk memastikan proporsi kelas tetap seimbang di setiap fold, mengingat dataset bersifat imbalanced. Setiap pipeline diuji menggunakan F2-score melalui cross_val_score(), sehingga diperoleh rata-rata F2 (mean), deviasi standar (std), serta performa masing-masing fold. Hasil ini digunakan untuk membandingkan model berdasarkan kemampuan mendeteksi churn secara konsisten dan stabil, sehingga model dengan performa terbaik dan variasi terendah dapat dipilih untuk tahap evaluasi akhir.

```python
f2_mean = []
f2_std = []
all_f2 = []
model_name = []
resampler_name = []

# Mencari model terbaik berdasarkan rerata f2 score terkecil dan std terkecil
for algoritma in models:
    for resampler in resamplers:

        model_pipe = ImbPipeline([
            ('preprocessing', prepros), 
            ('resampler', resampler),
            ('model', algoritma)
        ])

        skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

        model_cv = cross_val_score(
            model_pipe,                 # model yang sudah dimasukkan ke dalam pipeline
            xtrain,                    # data sebelum di preprocessing
            ytrain, 
            cv = skfold, 
            scoring = f2_scorer,
            error_score='raise'
            )

        f2_mean.append(model_cv.mean())
        f2_std.append(model_cv.std())
        all_f2.append(model_cv.round(4))
        model_name.append(algoritma.__class__.__name__)
        resampler_name.append(resampler.__class__.__name__)
```

```python
cv_results_df = pd.DataFrame({
    'model': model_name,
    'resampler': resampler_name,
    'mean f2 (cv)': f2_mean,
    'std': f2_std
}).sort_values(['mean f2 (cv)','std'], ascending=False)
cv_results_df.head(10)
```

```python
# Evaluate the models on the test set
test_f2_scores = []
model_names_test = []
resampler_names_test = []

for algoritma in models:
    for resampler in resamplers:
        model_pipe = ImbPipeline([
            ('preprocessor', prepros),
            ('resampler', resampler),
            ('model', algoritma)
        ])

        # Fitting the model
        model_pipe.fit(xtrain, ytrain)

        # Predicting
        y_pred = model_pipe.predict(xtest)

        # Evaluate with F2 score
        test_f2 = fbeta_score(ytest, y_pred, beta=2)
        test_f2_scores.append(test_f2)
        model_names_test.append(algoritma.__class__.__name__)
        resampler_names_test.append(resampler.__class__.__name__)
```

```python
# Store test set results in a DataFrame
test_results_df = pd.DataFrame({
    'model': model_names_test,
    'resampler': resampler_names_test,
    'f2 (test_set)': test_f2_scores
}).sort_values('f2 (test_set)', ascending=False)


test_results_df.head(10)
```

```python
merged_df = pd.merge(cv_results_df, test_results_df, on=['model', 'resampler']).drop(columns=['std'])
merged_df['diff'] = merged_df['f2 (test_set)'] - merged_df['mean f2 (cv)']
pd.set_option('display.precision', 10)

merged_df.sort_values(['f2 (test_set)'], ascending=False).head(10)
```

* Sebagian besar model menunjukkan peningkatan F2 score pada test set, yang terlihat dari nilai *diff* yang seluruhnya bernilai positif. Ini mengindikasikan bahwa performa model cenderung lebih baik saat diuji pada data baru.

* Rentang nilai *diff* berada pada kisaran sekitar **0.020 hingga 0.057**, yang merupakan variasi wajar akibat pembagian data dan tidak menunjukkan adanya indikasi overfitting yang ekstrem.

* Perbedaan yang kecil antara F2 score CV dan test set juga menandakan bahwa model stabil dan tidak sensitif terhadap perubahan data.

* Strategi resampling memberikan dampak yang berbeda pada tiap model, tetapi pola umumnya menunjukkan bahwa **RandomOverSampler dan SMOTE menghasilkan performa lebih baik dibandingkan RandomUnderSampler**, terutama pada model boosting.

### **Model Evaluation**

Setelah melakukan data training/modelling, maka selanjutnya adalah dilakukan model evaluation berdasarkan best model yang telah dipilih melalui proses benchmarking. Model yang sudah disempurnakan kemudian diuji kembali menggunakan data yang benar-benar baru, sehingga dapat dilihat apakah model mampu mengenali pola churn secara konsisten dan dapat diandalkan.
```python
# Define hyperparameter XGBoost
xgb_param_grid = {
    'model__n_estimators': [200, 300],
    'model__max_depth': [4, 6, 8],
    'model__learning_rate': [0.05, 0.1],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0]
}

# Create pipeline
xgb_smote_pipe = ImbPipeline([
    ('preprocessing', prepros),
    ('resampler', smote),
    ('model', xgboost)
])

# Untuned XGB
xgb_smote_untuned = xgb_smote_pipe.fit(xtrain, ytrain)
y_pred_xgb_smote_untuned = xgb_smote_untuned.predict(xtest)

xgb_smote_untuned_f2 = fbeta_score(ytest, y_pred_xgb_smote_untuned, beta=2)
xgb_smote_untuned_recall = recall_score(ytest, y_pred_xgb_smote_untuned)
xgb_smote_untuned_precision = precision_score(ytest, y_pred_xgb_smote_untuned)

# Tuned XGB dengan GridSearch
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

xgb_smote_grid = GridSearchCV(
    estimator=xgb_smote_pipe,
    param_grid=xgb_param_grid,
    scoring=f2_scorer,
    cv=cv,
    n_jobs=-1,
    verbose=1
)

xgb_smote_grid.fit(xtrain, ytrain)

xgb_smote_best = xgb_smote_grid.best_estimator_
y_pred_xgb_smote_tuned = xgb_smote_best.predict(xtest)

xgb_smote_tuned_f2 = fbeta_score(ytest, y_pred_xgb_smote_tuned, beta=2)
xgb_smote_tuned_recall = recall_score(ytest, y_pred_xgb_smote_tuned)
xgb_smote_tuned_precision = precision_score(ytest, y_pred_xgb_smote_tuned)

print("XGB best params:", xgb_smote_grid.best_params_)
print("XGB best F2 (CV):", xgb_smote_grid.best_score_)
print("XGB F2 (test, tuned):", xgb_smote_tuned_f2)
```
